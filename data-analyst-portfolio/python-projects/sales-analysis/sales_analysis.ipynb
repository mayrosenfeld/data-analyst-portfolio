```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mayrosenfeld/data-analyst-portfolio/refs/heads/main/data-analyst-portfolio/python-projects/sales-analysis/data/tips.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
df.describe()
```

```python
df["total_bill"].sum()
```

```python
sales_by_day = df.groupby("day")["total_bill"].sum()
sales_by_day
```

```python
sales_by_day.plot(kind="bar")
plt.title("Total Sales by Day")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

Sales are highest on weekends (Saturday and Sunday), which suggests higher customer activity during those days.

```python
sales_by_time = df.groupby("time")["total_bill"].sum()
sales_by_time
```

```python
sales_by_time.plot(kind="bar")
plt.title("Sales by Time of Day")
plt.show()
```

```python
df.groupby("size")["total_bill"].mean()
```

```python
df["tip_percentage"] = (df["tip"] / df["total_bill"]) * 100
df.head()
```

```python
df.groupby("day")["tip_percentage"].mean()
```

Sales are highest on weekends, especially Saturday, showing peak business activity.

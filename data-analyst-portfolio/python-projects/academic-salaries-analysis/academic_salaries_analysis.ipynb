```python
import pandas as pd

url = "https://raw.githubusercontent.com/mayrosenfeld/data-analyst-portfolio/refs/heads/main/data-analyst-portfolio/python-projects/academic-salaries-analysis/data/Salaries.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
df.describe()
df.columns
df.isnull().sum()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

df.head()
```

```python
df.info()
df.describe()
df['rank'].value_counts()
df['sex'].value_counts()
df['discipline'].value_counts()
```

```python
avg_salary_rank = df.groupby('rank')['salary'].mean()
avg_salary_rank
```

```python
avg_salary_discipline = df.groupby('discipline')['salary'].mean()
avg_salary_discipline
```

```python
avg_salary_sex = df.groupby('sex')['salary'].mean()
avg_salary_sex
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['yrs.since.phd'], df['salary'])
plt.xlabel('Years Since PhD')
plt.ylabel('Salary')
plt.title('Salary vs Years Since PhD')
plt.show()
```

Professors with higher rank earn more, as expected.
